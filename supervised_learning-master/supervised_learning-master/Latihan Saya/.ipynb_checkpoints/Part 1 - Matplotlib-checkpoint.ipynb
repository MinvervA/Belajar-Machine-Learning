```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

```python
X
```

```python
y
```

# Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X,y)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X,y1)
ax.plot(X,y2)
```

### Shortcut untuk visualisasi

```python
plt.figure(figsize=(6,6))
plt.plot(X,y1,'r-',label='sin')
plt.plot(X,y2,'g-',label='cos')
plt.xlim(0,10)
plt.ylim(-2,2)
plt.title('SIN & COS')
plt.xlabel('Xlabel')
plt.ylabel('Ylabel')
plt.legend()
```

```python
plt.figure(figsize=(6,6))
plt.plot(X,y1,'r-')
plt.plot(X,y2,'g-')
plt.axis('equal')
```

# Histogram

```python
numbers = np.random.rand(100000)
plt.hist(numbers,bins=25);
```

```python
numbers = np.random.randn(100000)
plt.hist(numbers,bins=80,density=1);
```

```python
numbers = np.random.randn(100000)
plt.hist(numbers,bins=80,density=1,cumulative=True);
```

# Subplot

### MATLAB Style

```python
plt.subplot(2,3,1)
plt.hist(numbers,bins=80,density=1);

plt.subplot(2,3,3)
plt.hist(numbers,bins=80,density=1);

plt.subplot(2,3,5)
plt.hist(numbers,bins=80,density=1);

plt.subplot(2,3,6)
plt.hist(numbers,bins=80,density=1);
```

### Matplotlib Style

```python
fig, ax =plt.subplots(2,3,figsize=(15,10))
ax[0,0].hist(numbers,bins=80,density=1);
ax[0,2].hist(numbers,bins=80,density=1);
ax[1,1].hist(numbers,bins=80,density=1);
ax[1,2].hist(numbers,bins=80,density=1);
```

```python
fig, ax =plt.subplots(2,3,figsize=(15,10))
ax[0,0].hist(numbers,bins=80,density=1);
# kalau mau dimatikan
ax[0,1].axis('off')
ax[0,2].hist(numbers,bins=80,density=1);
ax[1,1].hist(numbers,bins=80,density=1);
ax[1,2].hist(numbers,bins=80,density=1);
```

# Text dan Annotation

```python
numbers = np.random.randn(100000)
```

```python
plt.figure(figsize=(6,6))
plt.hist(numbers,bins=80,density=1);
plt.ylim(0,1)
# membuat garis merah ditengah
plt.axvline(0, color='r')
# untuk menambahkan text
plt.text(0, 0.8,'mean=0',ha='center')
# untuk menambah panah
plt.annotate('mean=0', xy=(0, 0.4), xytext=(1, 0.5), ha='center', arrowprops={'arrowstyle': '->'})
```

# 3D plot

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
```

```python
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

X, Y = np.meshgrid(x,y)
z = np.sqrt(X**2 + Y**2 + 10)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z,antialiased=False)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z,antialiased=False,cmap='winter')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z,antialiased=False,cmap='winter')
ax.view_init(elev=10)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, z,)
```
