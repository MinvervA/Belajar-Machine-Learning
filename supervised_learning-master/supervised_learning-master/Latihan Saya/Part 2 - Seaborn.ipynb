# Seaborn & Pandas Dataframe

## Statistical Summary

```python
df = pd.read_csv("../03 - Matplotlib Intro/data/cereal.csv")
df.head()
```

## Missing Value

```python
plt.figure(figsize=(6,6))
sns.heatmap(df.isna(), cmap='Blues')
```

## Correlation Heatmap

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.select_dtypes(include='number').corr(), cmap='bwr',cbar=False,vmin=-1,vmax=1,annot=True, fmt=".1f")
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="shelf",vars=["calories","fat","rating"], kind="reg")
```

```python
sns.jointplot(x="potass", y="fiber", data=df)
```

```python
sns.jointplot(x="potass", y="fiber", data=df, marginal_kws={"bins":25}, kind="reg")
```

# Categorical Data

## Persebaran

```python
sns.catplot(
    x="mfr",
    y="rating",
    data=df,
    kind="bar"
)
```

```python
sns.boxplot(x="mfr",y="rating",data=df)
```

```python
sns.violinplot(x="mfr",y="rating", data=df)
```

## Count

```python
sns.catplot(x="mfr", data=df, kind="count", hue="shelf")
```
